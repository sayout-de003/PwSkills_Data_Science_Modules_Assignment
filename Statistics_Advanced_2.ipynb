**Question 1: What is hypothesis testing in statistics?**

**Answer:**

Hypothesis testing is a statistical method used to make inferences about a population parameter based on sample data. It involves formulating a hypothesis about the population, collecting data, and then using statistical tests to determine whether there is enough evidence to support or reject the hypothesis. The goal is to assess the likelihood that the observed data occurred by chance if the hypothesis were true.

**Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?**

**Answer:**

The **null hypothesis (H₀)** is a statement that there is no significant difference or relationship between the variables being tested. It represents the status quo or the default assumption. In hypothesis testing, we aim to find evidence to reject the null hypothesis.

The **alternative hypothesis (H₁ or Ha)** is a statement that contradicts the null hypothesis. It proposes that there is a significant difference or relationship between the variables. This is the hypothesis the researcher is trying to prove.

The key difference lies in their starting point and goal:

*   **Null Hypothesis (H₀):** Assumes no effect or difference. The goal is to see if the data provides enough evidence to reject this assumption.
*   **Alternative Hypothesis (H₁):** Assumes there is an effect or difference. The goal is to find evidence to support this claim.

**Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.**

**Answer:**

The **significance level (α)**, also known as the alpha level, is a probability threshold that is set *before* conducting a hypothesis test. It represents the maximum probability of rejecting the null hypothesis when it is actually true (committing a Type I error).

Its role in deciding the outcome of a test is crucial:

*   We compare the p-value obtained from our statistical test to the significance level.
*   If the p-value is less than or equal to the significance level (p ≤ α), we reject the null hypothesis. This suggests that the observed data is unlikely to have occurred by random chance if the null hypothesis were true, providing evidence in favor of the alternative hypothesis.
*   If the p-value is greater than the significance level (p > α), we fail to reject the null hypothesis. This means that the observed data is reasonably likely to have occurred by random chance even if the null hypothesis were true, and there is not enough evidence to support the alternative hypothesis.

Commonly used significance levels are 0.05 (5%), 0.01 (1%), and 0.10 (10%). The choice of significance level depends on the context of the research and the consequences of making a Type I error. A smaller significance level makes it harder to reject the null hypothesis, reducing the risk of a Type I error but increasing the risk of a Type II error (failing to reject the null hypothesis when it is false).

**Question 4: What are Type I and Type II errors? Give examples of each.**

**Answer:**

In hypothesis testing, we aim to make a decision about the null hypothesis based on sample data. However, there's always a possibility of making an incorrect decision. These incorrect decisions are known as Type I and Type II errors.

*   **Type I Error (False Positive):** A Type I error occurs when you **reject the null hypothesis (H₀)** when it is actually **true**. In simpler terms, you conclude there is a significant effect or difference when there isn't one in the population.

    *   **Alpha (α):** The probability of making a Type I error is denoted by the significance level, α.
    *   **Example:** A medical test for a disease incorrectly indicates that a healthy person has the disease. Here, the null hypothesis is that the person is healthy (no disease), and the test rejects this null hypothesis, leading to a false positive.

*   **Type II Error (False Negative):** A Type II error occurs when you **fail to reject the null hypothesis (H₀)** when it is actually **false**. In simpler terms, you fail to detect a significant effect or difference that actually exists in the population.

    *   **Beta (β):** The probability of making a Type II error is denoted by β. The power of a test (1 - β) is the probability of correctly rejecting a false null hypothesis.
    *   **Example:** A medical test for a disease incorrectly indicates that a person who has the disease is healthy. Here, the null hypothesis is that the person is healthy (no disease), and the test fails to reject this null hypothesis, leading to a false negative.

It's important to understand the trade-off between Type I and Type II errors. Decreasing the probability of a Type I error (by lowering α) typically increases the probability of a Type II error (increasing β), and vice versa, assuming the sample size is fixed. The choice of significance level depends on the relative costs and consequences of making each type of error in a particular context.

**Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.**

**Answer:**

Both Z-tests and T-tests are used to compare means. The main difference is when to use them:

*   **Z-test:** Used when you know the **population standard deviation** and usually with a **large sample size** (n ≥ 30). It uses the Z-distribution.
*   **T-test:** Used when you **don't know the population standard deviation** and usually with a **small sample size** (n < 30). It uses the t-distribution.

In short, use a Z-test when you have more information about the population (known standard deviation), and a T-test when you have less information (unknown standard deviation, estimated from the sample).

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the binomial distribution
n = 10  # number of trials
p = 0.5  # probability of success

# Generate random numbers from a binomial distribution
binomial_data = np.random.binomial(n, p, 1000)  # generate 1000 data points

# Plot the histogram
plt.hist(binomial_data, bins=np.arange(-0.5, n + 1.5, 1), density=True, alpha=0.7, color='skyblue', edgecolor='black')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xticks(np.arange(0, n + 1, 1))
plt.grid(axis='y', alpha=0.75)
plt.show()
```

**Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.**

```python
import numpy as np
from scipy import stats

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Population parameters
population_mean = 50
population_std = 0.8  # known population std
sample_size = len(sample_data)

# Sample mean
sample_mean = np.mean(sample_data)

# Z-statistic
z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Two-tailed p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print(f"Sample Mean = {sample_mean:.3f}")
print(f"Z-Statistic = {z_stat:.3f}")
print(f"P-Value = {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Sample mean is significantly different from population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference from population mean.")
```

Question 8: Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate data from a normal distribution
np.random.seed(42)  # for reproducibility
mu = 50        # true mean
sigma = 5      # true standard deviation
n = 100        # sample size

data = np.random.normal(loc=mu, scale=sigma, size=n)

# Step 2: Calculate sample mean and standard error
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # sample standard deviation
se = sample_std / np.sqrt(n)       # standard error

# Step 3: Calculate 95% confidence interval
confidence = 0.95
z_score = stats.norm.ppf((1 + confidence) / 2)
ci_lower = sample_mean - z_score * se
ci_upper = sample_mean + z_score * se

print(f"Sample Mean = {sample_mean:.2f}")
print(f"95% Confidence Interval = ({ci_lower:.2f}, {ci_upper:.2f})")

# Step 4: Plot the data
plt.figure(figsize=(10, 5))
plt.hist(data, bins=15, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label='Sample Mean')
plt.axvline(ci_lower, color='green', linestyle='dashed', linewidth=2, label='95% CI Lower')
plt.axvline(ci_upper, color='green', linestyle='dashed', linewidth=2, label='95% CI Upper')
plt.title('Histogram of Simulated Normal Data with 95% Confidence Interval')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

**Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def calculate_and_visualize_zscores(data):
    """
    Calculates the Z-scores for a dataset, visualizes the standardized data
    using a histogram, and explains the meaning of Z-scores.

    Args:
        data (list or np.array): The input dataset.
    """
    # Convert data to numpy array
    data = np.array(data)

    # Calculate mean and standard deviation of the data
    mean = np.mean(data)
    std_dev = np.std(data) # By default, np.std calculates population standard deviation

    # Calculate Z-scores
    z_scores = (data - mean) / std_dev

    print(f"Original Data Mean: {mean:.4f}")
    print(f"Original Data Standard Deviation: {std_dev:.4f}")
    print(f"Calculated Z-scores (first 10): {z_scores[:10]}") # Print first 10 Z-scores

    # Visualize the standardized data (Z-scores)
    plt.figure(figsize=(8, 6))
    plt.hist(z_scores, bins=20, density=True, alpha=0.7, color='salmon', edgecolor='black')
    plt.xlabel('Z-score')
    plt.ylabel('Density')
    plt.title('Histogram of Z-scores (Standardized Data)')
    plt.grid(axis='y', alpha=0.75)
    plt.show()

    # Explanation of Z-scores
    print("\nExplanation of Z-scores:")
    print("--------------------------")
    print("A Z-score measures how many standard deviations away from the mean a particular data point is.")
    print("It is calculated using the formula: Z = (X - μ) / σ, where:")
    print("X is the individual data point")
    print("μ (mu) is the mean of the dataset")
    print("σ (sigma) is the standard deviation of the dataset")
    print("\nInterpretation:")
    print("- A Z-score of 0 means the data point is exactly at the mean.")
    print("- A positive Z-score indicates the data point is above the mean.")
    print("- A negative Z-score indicates the data point is below the mean.")
    print("- The magnitude of the Z-score tells you how many standard deviations away from the mean the point is.")
    print("  For example, a Z-score of 2 means the data point is 2 standard deviations above the mean.")
    print("  A Z-score of -1.5 means the data point is 1.5 standard deviations below the mean.")
    print("\nZ-scores are useful for standardizing data from different distributions, making it easier to compare data points.")
    print("When data is standardized to Z-scores, it has a mean of 0 and a standard deviation of 1.")


# Example usage with a sample dataset
sample_dataset = [10, 12, 12, 13, 15, 17, 18, 20, 22, 25, 28, 30, 32, 35, 38, 40]
calculate_and_visualize_zscores(sample_dataset)

# Example usage with the normal data from the previous question
# calculate_and_visualize_zscores(normal_data) # Uncomment to use data from Q8
```
